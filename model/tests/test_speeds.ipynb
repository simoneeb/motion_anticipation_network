```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
v = 2 # mm/s
v_range = np.arange(0.1,2,0.1)
x_rf = 0.5 # mm


# t_peak = np.exp(-v_range/5)*0.12 +0.02
t_peak = 0.4/v_range+0.1
t_peak = v_range
#t_peak = -10*v_range + 5
```

```python
def x_to_t(x,v):
    return x/v 


def t_to_x(t,v):
    return t*v


def anticipation(bar,cell):
    return bar-cell
```

```python
t_bar = x_to_t(x_rf,v_range)
```

```python

plt.plot(v_range,t_bar, label = f'bar at x = {x_rf} mm')
#plt.plot(v_range,t_peak, label = 'cell peaks')
plt.legend()
plt.xscale('log')
#plt.ylim(0,6)
```

```python
x_peak = t_to_x(t_peak,v_range)
```

```python
plt.axhline(x_rf, label = 'rf center position')
plt.plot(v_range,x_peak, label = 'cell peaks')
plt.legend()
```

```python
plt.axhline(0, label = 'rf center position')
plt.plot(v_range,anticipation(x_rf,x_peak), label = 'anticipation')
plt.legend()
```

```python

plt.axhline(0, label = 'rf center position')
plt.plot(v_range,anticipation(x_rf,x_peak), label = 'anticipation space')
plt.plot(v_range,anticipation(t_bar,t_peak), label = 'anticipation time')
plt.legend()
```
